```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_stadiums_in_India"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
# Find all td elements
td_elements = soup.find_all("td")
# td_elements[3].get_text()
td_elements[1]
```

```python
# Extract text from each td element
list_general=[]
for td in td_elements:
    text = td.get_text()
    list_general.append(text)
list_general
```

```python
# The data we want to extract starts with name: Narendra Modi Stadium 
# The last name in the table is : East Coast Railway Stadium
search_first_row_text="Narendra Modi Stadium"
search_last_row_text="East Coast Railway Stadium"
```

```python
# let us find their index in the list
search_index_first_row=list_general.index(search_first_row_text)
search_index_last_row=list_general.index(search_last_row_text)

```

```python
# Extracting Table-Only content from the above list
Stadium=[]
for i in range(search_index_first_row,search_index_last_row+1,7):
    Stadium.append(list_general[i])
print(Stadium)
```

```python
# Extracting Table-Only content from the above list
Capacity=[]
for i in range(search_index_first_row+1,search_index_last_row+2,7):
    Capacity.append(list_general[i])
print(Capacity)
```

```python
# Extracting Table-Only content from the above list
Games=[]
for i in range(search_index_first_row+2,search_index_last_row+3,7):
    Games.append(list_general[i])
print(Games)
```

```python
# Extracting Table-Only content from the above list
City=[]
for i in range(search_index_first_row+3,search_index_last_row+4,7):
    City.append(list_general[i])
print(City)
```

```python
# Extracting Table-Only content from the above list
State=[]
for i in range(search_index_first_row+4,search_index_last_row+5,7):
    State.append(list_general[i])
print(State)
```

```python
# Extracting Table-Only content from the above list
Home_team=[]
for i in range(search_index_first_row+5,search_index_last_row+6,7):
    Home_team.append(list_general[i])
print(Home_team)
```
